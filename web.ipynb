```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import orjson

ios = "https://support.apple.com/en-in/guide/iphone/toc/18.0"

resp = requests.get(ios, timeout=20)
resp.raise_for_status()

soup = BeautifulSoup(resp.text, "html.parser")

links = set()

for a in soup.find_all("a", href=True):
    href = urljoin(ios, a["href"])
    links.add(href)

print(len(links))
print(list(links)[:10])
```

```python
import orjson
```

```python
with open("ios_data.json", "wb") as f:
    f.write(orjson.dumps(list(links), option=orjson.OPT_INDENT_2))
```

```python
ios18_links = [u for u in links if "/ios/18.0" in u]

len(ios18_links)
```

```python
with open("ios_18_data.json", "wb") as f:
    f.write(orjson.dumps(ios18_links, option=orjson.OPT_INDENT_2))
```
